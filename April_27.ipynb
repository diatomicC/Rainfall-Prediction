# Merging all separate datasets

```python
import pandas as pd

# Load the CSV files
path_to_rainfall = '/Users/limengyu/SEEM 3650/Project/Rainfall.csv'
path_to_windspeed = '/Users/limengyu/SEEM 3650/Project/WindSpeed.csv'
path_to_cloudamount = '/Users/limengyu/SEEM 3650/Project/CloudAmount.csv'
path_to_humidty = '/Users/limengyu/SEEM 3650/Project/Humidity.csv'
path_to_sun = '/Users/limengyu/SEEM 3650/Project/Sun.csv'
path_to_maxtemp = '/Users/limengyu/SEEM 3650/Project/MaximumTemperature.csv'
path_to_mintemp = '/Users/limengyu/SEEM 3650/Project/MinimumTemperature.csv'
path_to_meantemp = '/Users/limengyu/SEEM 3650/Project/MeanTemperature.csv'

rainfall = pd.read_csv(path_to_rainfall)
windspeed = pd.read_csv(path_to_windspeed)
cloudamount = pd.read_csv(path_to_cloudamount)
humidity = pd.read_csv(path_to_humidty)
sun = pd.read_csv(path_to_sun)
maxtemp = pd.read_csv(path_to_maxtemp)
mintemp = pd.read_csv(path_to_mintemp)
meantemp = pd.read_csv(path_to_meantemp)

# Merge the dataframes on the 'Year', 'Month', and 'Day' columns
from functools import reduce
data_frames = [rainfall, windspeed, cloudamount, humidity, sun, maxtemp, mintemp, meantemp]
merged_data = reduce(lambda left, right: pd.merge(left, right, on=['Year', 'Month', 'Day']), data_frames)

# Convert 'Year', 'Month', 'Day' to a 'Date' column
merged_data['Date'] = pd.to_datetime(merged_data[['Year', 'Month', 'Day']])

# Drop the original 'Year', 'Month', 'Day' columns
merged_data.drop(['Year', 'Month', 'Day'], axis=1, inplace=True)

# Reorder columns to match requested format and keep the 'Date' column
column_order = ['Date', 'Total Rainfall (mm)', 'Mean Wind Speed (km/h)', 'Mean Amount of Cloud (%)','Humidity','Sun', 'Maximum Temperature', 'Minimum Temperature', 'Mean Temperature']
merged_data = merged_data[column_order]

# Save the merged data to a new CSV file
merged_data.to_csv('/Users/limengyu/SEEM 3650/Project/Training.csv', index=False)

print("Data has been successfully merged and saved.")
```

# Cardinality check for Categorical features:
Cardinality: The number of unique values in each categorical feature is known as cardinality.
A feature with a high number of distinct/ unique values is a high cardinality feature. A categorical feature with hundreds of zip codes is the best example of a high cardinality feature.
This high cardinality feature poses many serious problems like it will increase the number of dimensions of data when that feature is encoded. This is not good for the model.

```python
unique_values = len(rain['Total Rainfall (mm)'].unique())
print("Cardinality(no. of unique values) of 'Total Rainfall (mm)' are: {}".format(unique_values))
```

```python
# Get the data for two days before and after a given date, and calculate the mean and median corresponding to these dates
def calculate_mean_median(df, date, attribute):
    df['Date'] = pd.to_datetime(df['Date'])
    
    target_date = pd.to_datetime(date)
    date_range = pd.date_range(start=target_date - pd.DateOffset(days=2), 
                               end=target_date + pd.DateOffset(days=2))
    
    date_range = date_range[date_range != target_date]
    
    surrounding_data = df[df['Date'].isin(date_range)]
    
    mean_value = surrounding_data[attribute].mean()
    median_value = surrounding_data[attribute].median()
    
    return mean_value, median_value

mean, median = calculate_mean_median(rain, '2022-06-12', 'Maximum Temperature')
print(f"Mean for Maximum Temperature: {mean}")
print(f"Median for Maximum Temperature: {median}")
print("\n") 

mean, median = calculate_mean_median(rain, '2022-06-12', 'Minimum Temperature')
print(f"Mean for Minimum Temperature: {mean}")
print(f"Median for Minimum Temperature: {median}")
print("\n") 

mean, median = calculate_mean_median(rain, '2022-06-12', 'Mean Temperature')
print(f"Mean for Mean Temperature: {mean}")
print(f"Median for Mean Temperature: {median}")
print("\n") 
print("\n") 

mean, median = calculate_mean_median(rain, '2022-06-15', 'Maximum Temperature')
print(f"Mean for Maximum Temperature: {mean}")
print(f"Median for Maximum Temperature: {median}")
print("\n") 

mean, median = calculate_mean_median(rain, '2022-06-15', 'Minimum Temperature')
print(f"Mean for Minimum Temperature: {mean}")
print(f"Median for Minimum Temperature: {median}")
print("\n") 

mean, median = calculate_mean_median(rain, '2022-06-15', 'Mean Temperature')
print(f"Mean for Mean Temperature: {mean}")
print(f"Median for Mean Temperature: {median}")
print("\n") 
```
