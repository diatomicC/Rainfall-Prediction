# Merging all separate datasets

```python
import pandas as pd

# Load the CSV files
path_to_rainfall = '/Users/limengyu/SEEM 3650/Project/Rainfall.csv'
path_to_windspeed = '/Users/limengyu/SEEM 3650/Project/WindSpeed.csv'
path_to_cloudamount = '/Users/limengyu/SEEM 3650/Project/CloudAmount.csv'
path_to_humidty = '/Users/limengyu/SEEM 3650/Project/Humidity.csv'
path_to_sun = '/Users/limengyu/SEEM 3650/Project/Sun.csv'
path_to_maxtemp = '/Users/limengyu/SEEM 3650/Project/MaximumTemperature.csv'
path_to_mintemp = '/Users/limengyu/SEEM 3650/Project/MinimumTemperature.csv'
path_to_meantemp = '/Users/limengyu/SEEM 3650/Project/MeanTemperature.csv'

rainfall = pd.read_csv(path_to_rainfall)
windspeed = pd.read_csv(path_to_windspeed)
cloudamount = pd.read_csv(path_to_cloudamount)
humidity = pd.read_csv(path_to_humidty)
sun = pd.read_csv(path_to_sun)
maxtemp = pd.read_csv(path_to_maxtemp)
mintemp = pd.read_csv(path_to_mintemp)
meantemp = pd.read_csv(path_to_meantemp)

# Merge the dataframes on the 'Year', 'Month', and 'Day' columns
from functools import reduce
data_frames = [rainfall, windspeed, cloudamount, humidity, sun, maxtemp, mintemp, meantemp]
merged_data = reduce(lambda left, right: pd.merge(left, right, on=['Year', 'Month', 'Day']), data_frames)

# Convert 'Year', 'Month', 'Day' to a 'Date' column
merged_data['Date'] = pd.to_datetime(merged_data[['Year', 'Month', 'Day']])

# Drop the original 'Year', 'Month', 'Day' columns
merged_data.drop(['Year', 'Month', 'Day'], axis=1, inplace=True)

# Reorder columns to match requested format and keep the 'Date' column
column_order = ['Date', 'Total Rainfall (mm)', 'Mean Wind Speed (km/h)', 'Mean Amount of Cloud (%)','Humidity','Sun', 'Maximum Temperature', 'Minimum Temperature', 'Mean Temperature']
merged_data = merged_data[column_order]

# Save the merged data to a new CSV file
merged_data.to_csv('/Users/limengyu/SEEM 3650/Project/Training.csv', index=False)

print("Data has been successfully merged and saved.")
```

```python
# Get the data for two days before and after a given date, and calculate the mean and median corresponding to these dates
def calculate_mean_median(df, date, attribute):
    df['Date'] = pd.to_datetime(df['Date'])
    
    target_date = pd.to_datetime(date)
    date_range = pd.date_range(start=target_date - pd.DateOffset(days=2), 
                               end=target_date + pd.DateOffset(days=2))
    
    date_range = date_range[date_range != target_date]
    
    surrounding_data = df[df['Date'].isin(date_range)]
    
    mean_value = surrounding_data[attribute].mean()
    median_value = surrounding_data[attribute].median()
    
    return mean_value, median_value

mean, median = calculate_mean_median(rain, '2022-06-12', 'Maximum Temperature')
print(f"Mean for Maximum Temperature: {mean}")
print(f"Median for Maximum Temperature: {median}")
print("\n") 

mean, median = calculate_mean_median(rain, '2022-06-12', 'Minimum Temperature')
print(f"Mean for Minimum Temperature: {mean}")
print(f"Median for Minimum Temperature: {median}")
print("\n") 

mean, median = calculate_mean_median(rain, '2022-06-12', 'Mean Temperature')
print(f"Mean for Mean Temperature: {mean}")
print(f"Median for Mean Temperature: {median}")
print("\n") 
print("\n") 

mean, median = calculate_mean_median(rain, '2022-06-15', 'Maximum Temperature')
print(f"Mean for Maximum Temperature: {mean}")
print(f"Median for Maximum Temperature: {median}")
print("\n") 

mean, median = calculate_mean_median(rain, '2022-06-15', 'Minimum Temperature')
print(f"Mean for Minimum Temperature: {mean}")
print(f"Median for Minimum Temperature: {median}")
print("\n") 

mean, median = calculate_mean_median(rain, '2022-06-15', 'Mean Temperature')
print(f"Mean for Mean Temperature: {mean}")
print(f"Median for Mean Temperature: {median}")
print("\n") 
```

# Outliers detection and treatment:

```python
mean = rain['Mean Temperature'].mean()
std_dev = rain['Mean Temperature'].std()

# Setting the cutoff to 3 standard deviations from the mean
cutoff = std_dev * 3

# Lower and upper bounds
lower_bound = mean - cutoff
upper_bound = mean + cutoff

# Filtering the data for outliers
outliers = rain[(rain['Mean Temperature'] < lower_bound) | (rain['Mean Temperature'] > upper_bound)]
print("Outliers using Standard Deviation Method:")
print(outliers)
```

```python
# Calculate Q1, Q3, and IQR
Q1 = rain['Mean Temperature'].quantile(0.25)
Q3 = rain['Mean Temperature'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter outliers
outliers = rain[(rain['Mean Temperature'] < lower_bound) | (rain['Mean Temperature'] > upper_bound)]
print("Outliers using IQR Method:")
print(outliers)
```

#  Model Building

```python
# We will use the Logistic Regression to build a predictive model to predict whether or not it will rain tomorrow in HK
# Model Training:
from sklearn.linear_model import LogisticRegression
classifier_logreg = LogisticRegression(solver='liblinear', random_state=0)
classifier_logreg.fit(X_train, y_train)
```

```python
# Model Testing:
y_pred = classifier_logreg.predict(X_test)
y_pred
```

# Evaluating Model Performance:

```python
# 1, accuracy_score()
from sklearn.metrics import accuracy_score
print("Accuracy Score: {}".format(accuracy_score(y_test,y_pred)))
```

```python
# Checking for Underfitting and Overfitting:
print("Train Data Score: {}".format(classifier_logreg.score(X_train, y_train)))
print("Test Data Score: {}".format(classifier_logreg.score(X_test, y_test)))
```

 The accuracy scores for both the training and testing datasets are similar. 
 This indicates that neither underfitting nor overfitting is a concern. 
 Therefore, the model is effectively generalizing to new, unseen data.

```python
# 2，Confusion Matrix:
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix:")
print(cm)
```

```python
# 3，Classification-report:
from sklearn.metrics import classification_report

report = classification_report(y_test, y_pred)

print("Classification Report:")
print(report)
```

```python
# The mean accuracy score of cross-validation is lower than the original model's accuracy score, which is 0.8737. 
# The model's accuracy may not improve using cross-validation, possibly due to limited data availability. 
# Our dataset is relatively small, with just over 1000 records, and cross-validation may not be sufficient to capture all aspects of the data stably. 
# With such a small dataset, the model's generalization capability might be limited because the training samples may not adequately represent the complexity of the overall data.
```

```python
# This could also be due to an imbalance in the data. 
# In our dataset, the number of samples in the "No Rain (0)" category far exceeds those in the "Rain (1)" category, 
# which may cause the model to be biased towards the majority class, resulting in a higher accuracy in the original model test. 
# Cross-validation, by more fairly evaluating the model's performance across different subsets, may reveal deficiencies in the model's handling of the less represented category.
```
