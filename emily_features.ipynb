```python
import pandas as pd

# Assuming df is your DataFrame
mean = df['Mean Temperature (°C)'].mean()
std_dev = df['Mean Temperature (°C)'].std()

# Setting the cutoff to 3 standard deviations from the mean
cutoff = std_dev * 3

# Lower and upper bounds
lower_bound = mean - cutoff
upper_bound = mean + cutoff

# Filtering the data for outliers
outliers = df[(df['Mean Temperature (°C)'] < lower_bound) | (df['Mean Temperature (°C)'] > upper_bound)]
print("Outliers using Standard Deviation Method:")
print(outliers)
```

```python
# Calculate Q1, Q3, and IQR
Q1 = df['Mean Temperature (°C)'].quantile(0.25)
Q3 = df['Mean Temperature (°C)'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter outliers
outliers = df[(df['Mean Temperature (°C)'] < lower_bound) | (df['Mean Temperature (°C)'] > upper_bound)]
print("Outliers using IQR Method:")
print(outliers)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('/Users/diatomicc/Desktop/SEEM3650/Training.csv')

# Convert relevant columns to numeric, if not already
columns = ['Total Rainfall (mm)', 'Mean Wind Speed (km/h)', 'Mean Amount of Cloud (%)', 'Sun', 'Humidity', 'Minimum Temperature (°C)', 'Mean Temperature (°C)', 'Maximum Temperature (°C)']
for column in columns:
    df[column] = pd.to_numeric(df[column], errors='coerce')

# Filling missing values with the mean for simplicity
df.fillna(df.mean(), inplace=True)

# Calculate Pearson correlation matrix
pearson_corr = df.corr(method='pearson')
print("Pearson Correlation Matrix:")
print(pearson_corr)

# Calculate Spearman correlation matrix
spearman_corr = df.corr(method='spearman')
print("\nSpearman Correlation Matrix:")
print(spearman_corr)
```

```python
# Using Seaborn to create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(pearson_corr, annot=True, fmt=".2f", cmap='coolwarm', cbar=True, square=True)
plt.title('Heatmap of Pearson Correlation Matrix')
plt.show()

plt.figure(figsize=(10, 8))
sns.heatmap(spearman_corr, annot=True, fmt=".2f", cmap='coolwarm', cbar=True, square=True)
plt.title('Heatmap of Spearman Correlation Matrix')
plt.show()
```

```python
df = pd.read_csv('/Users/diatomicc/Desktop/SEEM3650/Training.csv')

# Convert date column to datetime format for easier plotting
df['Date'] = pd.to_datetime(df['Date'])

# Convert 'Total Rainfall (mm)' to numeric, handling non-numeric values
df['Total Rainfall (mm)'] = pd.to_numeric(df['Total Rainfall (mm)'], errors='coerce')

# Filling missing values with 0 for simplicity, assuming days with missing data had no recorded rainfall
df['Total Rainfall (mm)'].fillna(0, inplace=True)
```

```python
# Extract month from the date
df['Month'] = df['Date'].dt.month

# Aggregate rainfall by month
monthly_rainfall = df.groupby('Month')['Total Rainfall (mm)'].sum()

# Alternatively, you can use mean instead of sum to see average monthly rainfall:
# monthly_rainfall = df.groupby('Month')['Total Rainfall (mm)'].mean()
```

```python
# Setting the plot size
plt.figure(figsize=(10, 6))

# Creating the bar chart
sns.barplot(x=monthly_rainfall.index, y=monthly_rainfall.values, palette="viridis")

# Adding titles and labels
plt.title('Total Rainfall per Month')
plt.xlabel('Month')
plt.ylabel('Total Rainfall (mm)')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# Display the plot
plt.show()
```
